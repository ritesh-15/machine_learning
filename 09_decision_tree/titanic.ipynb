```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
df = pd.read_csv("titanic.csv")
df.keys()
```

```python
df = df.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis="columns")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
```

```python
df["sex_n"] = le_sex.fit_transform(df["Sex"])
df
```

```python
df.drop(["Sex"], inplace=True, axis="columns")
df
```

```python
inputs = df.drop(["Survived"], axis="columns")
inputs
```

```python
import math

median_age = math.floor(inputs.Age.median())

median_age
```

```python
new_ages = inputs["Age"].fillna(median_age)
inputs["Age"] = new_ages
```

```python
target = df["Survived"]
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, train_size=0.8)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
