```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snb
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.keys()
```

```python
plt.gray()
for i in range(0,4):
    plt.matshow(digits.images[i])
plt.show()
```

```python
df = pd.DataFrame(digits.data)
df
```

```python
df.head()
```

```python
df["target"] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(["target"],axis="columns"), digits.target, train_size=0.8)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=80) # n_estimators number of decision trees
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)

cm
```

```python
plt.figure(figsize=(10,7))
snb.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```
